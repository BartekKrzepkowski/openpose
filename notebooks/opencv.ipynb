```python
%matplotlib inline
import cv2
from openpose import pyopenpose as op
from matplotlib import pyplot as plt
```

# Image

```python
IMG_PATH = '../examples/media/COCO_val2014_000000000428.jpg'
```

## Body

```python
PARAMS = dict()
PARAMS["model_folder"] = "../models/"
```

```python
# Starting OpenPose
opWrapper = op.WrapperPython()
opWrapper.configure(PARAMS)
opWrapper.start()
```

```python
# Process Image
datum = op.Datum()
imageToProcess = cv2.imread(IMG_PATH)
datum.cvInputData = imageToProcess
opWrapper.emplaceAndPop(op.VectorDatum([datum]))
```

```python
gray = cv2.cvtColor(datum.cvOutputData, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.xticks([]), plt.yticks([])
plt.imshow(gray);
plt.show()
```

```python
print("Body keypoints: \n" + str(datum.poseKeypoints))
```

## HeatMaps

```python
PARAMS = dict()
PARAMS["model_folder"] = "../models/"
PARAMS["heatmaps_add_parts"] = True
PARAMS["heatmaps_add_bkg"] = True
PARAMS["heatmaps_add_PAFs"] = True
PARAMS["heatmaps_scale"] = 2
```

```python
# Starting OpenPose
opWrapper = op.WrapperPython()
opWrapper.configure(PARAMS)
opWrapper.start()
```

```python
# Process Image
datum = op.Datum()
imageToProcess = cv2.imread(IMG_PATH)
datum.cvInputData = imageToProcess
opWrapper.emplaceAndPop(op.VectorDatum([datum]))
```

```python
# Process outputs
outputImageF = (datum.inputNetData[0].copy())[0,:,:,:] + 0.5
outputImageF = cv2.merge([outputImageF[0,:,:], outputImageF[1,:,:], outputImageF[2,:,:]])
outputImageF = (outputImageF*255.).astype(dtype='uint8')
heatmaps = datum.poseHeatMaps.copy()
heatmaps = (heatmaps).astype(dtype='uint8')
```

```python
# Display Image
for counter in range(5):
    num_maps = heatmaps.shape[0]
    heatmap = heatmaps[counter, :, :].copy()
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    combined = cv2.addWeighted(outputImageF, 0.5, heatmap, 0.5, 0)
    gray = cv2.cvtColor(combined, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 10))
    plt.xticks([]), plt.yticks([])
    plt.imshow(gray);
    plt.show()
```
